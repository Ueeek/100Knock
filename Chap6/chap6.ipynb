```python
import re
from nltk.stem.porter import PorterStemmer as PS
#from stemming import porter
import subprocess
import xml.etree.ElementTree as ET
import os
```

# 50

```python
##java 8じゃないと動かないってよ
def nlp_lines():
    '''
    nlp.txtを1行づつ読み込むジェネレータ
    '''
    with open("./nlp.txt") as lines:
        #/s のまえがgroup1 
        pattern = re.compile(r'''
            (^.*?[\.|;|:|\?|!]
            )
            \s 
            ([A-Z].*
            )
            ''',re.MULTILINE+re.VERBOSE+re.DOTALL)
        
        for line in lines:
            line = line.strip()
            while len(line)>0:
                match = pattern.match(line)
                if match:
                    yield match.group(1) 
                    line = match.group(2)
                else:
                    yield line
                    line= ''
                    
```

```python
for line in nlp_lines():
    print(line)
```

```python
with open("./nlp.txt") as f:
    for line in f:
        print(line)
```

# 51

```python
def nlp_word():
    for line in nlp_lines():
        for word in line.split():
            yield word.rstrip('.,;:?!')
        yield '' #文の終わりに改行がいるから''を返して、print('')で\nってことにする
```

```python
for word in nlp_word():
    print(word)
```

# 52

```python
ps = PS()
for word in nlp_word():
    print("{}\t{}".format(word,ps.stem(word)))
```

```python
def parse_nlp():
    '''
    nlp.txtをStanford Core NLP で解析してxmlファイルへ出力
    '''
    if not os.path.exists("nlp.txt.xml"):
        subprocess.run(
        'java -cp "/usr/local/lib/stanford-corenlp-full-2018-10-05/*"'
            ' -Xmx2g'
            ' edu.stanford.nlp.pipeline.StanfordCoreNLP'
            ' -annotators tokenize,ssplit,pos,lemma,ner,parse,dcoref'
            ' -file ' + "nlp.txt.xml" + ' 2>parse.out',
            shell=True,     # shellで実行
            check=True      # エラーチェックあり
        )
        
```

```python
parse_nlp()
root = ET.parse("./nlp.txt.xml")
for word in root.iter('word'):
    print(word.text)
```
